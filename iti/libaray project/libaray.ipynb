```python
class Book_list:
    def __init__(self , book_name , author_name , price , Quantity):
        self.book_name = book_name
        self.author_name = author_name
        self.price = price
        self.quantity = Quantity

    def display_info(self):
        print(f"book name : {self.book_name}\nauthor_name : {self.author_name}\nprice : {self.price}\nquantity : {self.quantity}")
        print(20*"_")

    def to_dict(self):
        return{
            "book name" : self.book_name,
            "author_name" : self.author_name,
            "price" : self.price,
            "quantity" : int(self.quantity)
            
        }
        
```

```python
class LendedBooks:
    def __init__(self,book_name,user_name,start_date,end_date):
        self.book_name = book_name
        self.user_name = user_name
        self.start_date = start_date
        self.end_date = end_date

    def display_Lendinfo(self):
        print(f"book name : {self.book_name}\nuser_name : {self.user_name}\nstartdate : {self.start_date}\nenddate : {self.end_date}")
        print(20*"_")
    def lend_to_dict(self):
        return{
            "book name" : self.book_name,
            "user_name" : self.user_name,
            "startdate" : self.start_date,
            "enddate" : self.end_date
            
        }
        
```

```python
import json
```

```python
class MainMange:
    def __init__(self):
        self.books = []
        self.lended = []
        self.file_path_books = "Book_list.json"
        self.file_path_Lended = "lended_list.json"
        self.load_file_books()
        self.load_file_lended()


    def add_books(self, book_name , author_name , price , Quantity):
        for book in self.books:
            if book.book_name == book_name:
                print("the book is already added \n increase the quantity")
                return
        self.books.append(Book_list(book_name , author_name , price , Quantity))
        self.save_to_file_books()
        print(f"the book {book_name} is added")

    def display_All_books(self):
        if not self.books :
            print("no books founded")
        else :
            print(20*"*")
            for book in self.books:
                book.display_info()

    def delete_book(self,book_name):
        new_lst = []
        book_founded = False
        for book in self.books:
            if book.book_name != book_name:
                new_lst.append(book)
            else:
                book_founded = True

        if  book_founded :
            self.books = new_lst
            self.save_to_file_books()
            print(f"the process ended and the book {book_name} is removed")
        else :
            print("there is no book by this name to delete it ")


    def edit_Quantity(self, book_name, Quantity):  
        for book in self.books:
            if book.book_name == book_name:
                book.quantity += Quantity
                self.save_to_file_books()
                print(f"{book_name} quantity : {book.quantity}")  
                return
    
        print(f"{book_name} is not found")  



    def book_lending(self,book_name,user_name,startdate,enddate):
        book_founded = False
        for book in self.books:
            if book.book_name == book_name :
                book_founded = True
                if book.quantity > 0:
                    book.quantity -= 1
                    self.lended.append(LendedBooks(book_name,user_name,startdate,enddate))
                    self.save_to_file_lended()
                    self.save_to_file_books()
                    print(f"the book {book_name} lended to {user_name} form {startdate} to {enddate}")
                    return
                else :
                    print("the book founded but not avilabe now out of stock")
                    return
        if not book_founded :
            print("the book is not in our library")


    def return_books(self,book_name ,user_name):
        user_book_flag = False
        for book in self.lended:
            if book.book_name == book_name and book.user_name == user_name:
                user_book_flag = True
                self.lended.remove(book)
                self.save_to_file_lended()
                print("the book is returned")
                for book in self.books:
                    if book.book_name ==book_name:
                        book.quantity +=1
                        self.save_to_file_books()

        if not user_book_flag :
            print("no book by this name lended to this user")


    def save_to_file_books(self):
        new_lst = []
        for book in self.books:
            new_lst.append(book.to_dict())
        with open(self.file_path_books ,mode="w") as fd:
            json.dump(new_lst , fd)


    def save_to_file_lended(self):
        new_lst = []
        for book in self.lended:
            new_lst.append(book.lend_to_dict())
        with open(self.file_path_Lended ,mode="w") as fd:
            json.dump(new_lst , fd)


    def load_file_books(self):
        try:
            with open(self.file_path_books,mode="r") as fd:
                data = json.load(fd)
                for book in data:
                    self.books.append(Book_list(book["book name"],book["author_name"],book["price"],int(book["quantity"])))

        except(FileNotFoundError , json.JSONDecodeError):
            self.books = [] 


    def load_file_lended(self):
        try:
            with open(self.file_path_Lended,mode="r") as fd:
                data = json.load(fd)
                for book in data:
                    self.lended.append(LendedBooks(book["book name"],book["user_name"],book["startdate"],book["enddate"]))

        except(FileNotFoundError , json.JSONDecodeError):
            self.lended = [] 
```

```python
class FrontEnd:
    def __init__(self):
        self.main = MainMange()
    
    def display_menu(self):
        while True :
            
            print("\n\t\tlibaray Managment System\t\t")
            print("1) display all books : ")
            print("2) add new book : ")
            print("3) delete books : ")
            print("4) update quantity of a book : ")
            print("5) lend a book : ")
            print("6) return a lended  book : ")
            print("7) End The Program | Exit")
            print(20 * "*")
            try:
                choice = int(input("please Enter the Number : "))
            except :
                print("Please Enter a Numerical Value")
            else:    

                if choice == 1 :

                    self.main.display_All_books()

                elif choice == 2 :
                    book_name = input("please Enter Book Name")
                    author_name = input("please Enter author Name")
                    price = input("please Enter Book price")
                    quantity = input("please Enter Book quantity")
                    self.main.add_books(book_name,author_name,price,quantity)


                elif choice == 3 :
                    book_name = input("please Enter Book Name")
                    self.main.delete_book(book_name)
                   
                    
                elif choice == 4 :
                    book_name = input("please Enter Book Name")
                    quantity = int(input("please Enter Book quantity"))
                    self.main.edit_Quantity(book_name,quantity)


                elif choice == 5 :
                    book_name = input("please Enter Book Name")
                    user_name = input("please Enter USER Name")
                    startdate = input("please Enter STARTDATE")
                    enddate = input("please Enter Book ENDDATE")
                    self.main.book_lending(book_name,user_name,startdate,enddate)


                elif choice == 6 :
                    book_name = input("please Enter Book Name")
                    user_name = input("please Enter USER Name")

                    self.main.return_books(book_name , user_name)

                elif choice == 7 :
                    print("Exit Program .... ")
                    break
                
                else  :
                    print("Please Enter a Vailed Number")
```
