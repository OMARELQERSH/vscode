#### 4. Plot survival rate by gender and class (combined).

#### 6. How does the number of siblings/spouses affect survival?


```python
sep_per_sur = df.groupby("SibSp")["Survived"].mean().reset_index()
sep_per_sur
```

```python
sep_per_surcount = df.groupby("SibSp")["Survived"].count().reset_index()
sep_per_surcount
```

```python
sep_per_surcount = df.groupby("SibSp")["Survived"].agg(count = "count",rate = "mean").reset_index()
sep_per_surcount
```

```python
px.bar(sep_per_sur,x="SibSp",y="Survived")
```
